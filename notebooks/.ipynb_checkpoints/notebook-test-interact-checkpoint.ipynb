```python
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
```

# Mean vs Median

```python
from scipy.stats import exponnorm

def plot_mean_median(K):
    #data
    data = exponnorm.rvs(K, size=1000) #for histogram
    rv = exponnorm(K) #for pdf
    x = np.linspace(exponnorm.ppf(0.01, K), exponnorm.ppf(0.99, K), 100) #x for the pdf
    #plot
    fig,ax = plt.subplots(figsize=(10, 4))
    ax.hist(data, bins=30, normed=True)
    ax.plot(x, rv.pdf(x), 'k-')
    ax.axvline(np.mean(data), color='r', label='mean')
    ax.axvline(np.median(data), color='b', label='median')
    ax.legend(loc='upper right')
    ax.set_xlim(-4, 65)
    plt.show()

interact(plot_mean_median, K=(0.04, 10, 0.01))
```

```python
from scipy.stats import weibull_min

def plot_mean_median2(K):
    #data
    data = weibull_min.rvs(K, size=1000) #for histogram
    rv = exponnorm(K) #for pdf
    x = np.linspace(weibull_min.ppf(0.01, K), weibull_min.ppf(0.99, K), 100) #x for the pdf
    #plot
    fig,ax = plt.subplots(figsize=(10, 4))
    ax.hist(data, bins=30, normed=True)
    ax.plot(x, rv.pdf(x), 'k-')
    ax.axvline(np.mean(data), color='r', label='mean')
    ax.axvline(np.median(data), color='b', label='median')
    ax.legend(loc='upper right')
    plt.show()

interact(plot_mean_median2, K=(0.5, 4, 0.1))#K=(0.04, 10, 0.01))
```
